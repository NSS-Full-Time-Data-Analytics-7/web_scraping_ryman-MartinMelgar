```python
import pandas as pd
import operator
import time
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import datetime 
%matplotlib inline
from bs4 import BeautifulSoup as BS
import requests
from IPython.core.display import HTML
```

```python
Tables = []
for i in range(1,6):
    URL = f'https://ryman.com/events/list/?tribe_event_display=list&tribe_paged={i}'
    response = requests.get(URL)
    soup = BS(response.text)
    Artists = pd.DataFrame(list(soup.findAll('a',attrs={'class' : 'tribe-event-url'})))
    Artists.rename(columns={ Artists.columns[0]: "Artists" }, inplace = True)
    Artists = Artists.Artists.str.split(pat= '\t\t', expand = True)
    Artists.rename(columns={ Artists.columns[0]: "Delete" }, inplace = True)
    Artists.rename(columns={ Artists.columns[1]: "Split" }, inplace = True)
    Artists = Artists.Split.str.split(pat= '\t', expand = True)
    Artists.rename(columns={ Artists.columns[0]: "Artists" }, inplace = True)
    Artists.rename(columns={ Artists.columns[1]: "Drop" }, inplace = True)
    Artists = Artists.drop(columns = ['Drop'])
    time = pd.DataFrame(list(soup.findAll('time')))
    time.rename(columns={ time.columns[0]: "Date" }, inplace = True)
    time = time.Date.str.split(pat= ' at ', expand = True)
    time.rename(columns={ time.columns[0]: "Date" }, inplace = True)
    time.rename(columns={ time.columns[1]: "Time" }, inplace = True)
    Openers = pd.DataFrame(list(soup.findAll('span',attrs={'class' : 'opener'})))
    Openers.rename(columns={ Openers.columns[0]: "Openers" }, inplace = True)
    Openers.loc[Openers.Openers.str.contains("With|with|featuring")==False, 'Openers'] = "No Opener"
    Event_Table = pd.concat([Artists, time, Openers], axis =1).replace(np.nan, "No Opener")
    Tables.append(Event_Table)
```

```python
Master_List = Tables[0].append([Tables[1],Tables[2],Tables[3],Tables[4]])
Master_List = Master_List.reset_index(drop= True)
Master_List
```
